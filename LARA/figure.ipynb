```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
 
g10=pd.read_csv('generate_data/g10.csv')
g20=pd.read_csv('generate_data/g20.csv')
m10=pd.read_csv('generate_data/g10.csv')
m20=pd.read_csv('generate_data/g20.csv')
p10=pd.read_csv('generate_data/g10.csv')
p20=pd.read_csv('generate_data/g20.csv')
#
#......
#......
#x_axixtrain_pn_dislist()
 
#
def figure_score(g,m,p):
    plt.title('Result Analysis')
    plt.plot(g.iloc[:, 0].values.tolist(),g.iloc[:, 1].values.tolist(), color='green', label='nDCN')
    plt.plot(m.iloc[:, 0].values.tolist(),m.iloc[:, 1].values.tolist(), color='red', label='mean average accuracy')
    plt.plot(p.iloc[:, 0].values.tolist(),p.iloc[:, 1].values.tolist(), label='p')
    plt.legend() # 

    plt.xlabel('iteration times')
    plt.ylabel('score')
    plt.show()
print('g10,m10,p10')
figure_score(g10,m10,p10)
print('g20,m20,p20')
figure_score(g20,m20,p20)
```
